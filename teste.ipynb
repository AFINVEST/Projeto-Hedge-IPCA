```python
import pandas as pd
df_debentures = pd.read_csv("deb_table_completa2.csv")
df_debentures
```

```python
#Tenho que filtrar pelos ativos AESOA1 e HBSA11
df_debentures2 = df_debentures[df_debentures['Ativo'].isin(['AESOA1', 'HBSA11'])]
df_debentures2
```

```python
#Ver se tem ativos duplicados
ativos_duplicados = df_debentures2[df_debentures2.duplicated(subset=["Ativo", "Data de pagamento"], keep="first")]
ativos_duplicados.info()
```

```python
df_debentures
```

```python
# 1. Identificar duplicatas apenas para os ativos desejados
duplicatas_especificas = df_debentures[
    df_debentures['Ativo'].isin(['AESOA1', 'HBSA11'])
].duplicated(subset=["Ativo", "Data de pagamento"], keep='first')

# 2. Criar uma máscara booleana geral para o dataframe inteiro
mascara_geral = df_debentures['Ativo'].isin(['AESOA1', 'HBSA11']) & duplicatas_especificas

# 3. Remover essas linhas duplicadas
df_debentures = df_debentures[~mascara_geral]
df_debentures
```

```python
#Salvar novamente
df_debentures.to_csv("deb_table_completa2.csv", index=False)
```
