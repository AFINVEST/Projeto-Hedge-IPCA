```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/deb_table_completa3.csv')
dados
```

```python
dados_faltantes = pd.read_csv('Dados/deb_table_faltantes.csv')
dados_faltantes
```

```python
#Adicionando os dados faltantes ao dataframe original
dados = pd.concat([dados, dados_faltantes], ignore_index=True)
dados
```

```python
# Ver se tem linhas que estão vazias em todas as colunas menos na coluna Ativo para identificar quais ativos tiveram erro ao ser capturados
dados[dados.iloc[:, :-1].isnull().all(axis=1)]
```

```python
# Tirar as linhas que estão vazias em todas as colunas menos na coluna Ativo
dados = dados[~dados.iloc[:, :-1].isnull().all(axis=1)]
dados
```
